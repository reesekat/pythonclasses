Loops-

A loop helps us repeat the same task again and again without writing the same code many times

example-
writing attendance for 30 students
printing numbers from 1 to 100
checking marks for each subject

Types of loops-
for loop
while loop

FOR LOOP-

A for loop takes one item at a time from a collection and works on it.

syntax-
for item in collection:
    print(item)

collection-list/string/range
item-temporary variable

```python
#basic example

for p in range(5):   
    print(p)

#range(5) ->numbers from 0 to 4
#i takes one value at a time
#print(i) runs repeatedly until the for loop condition is true
```

```python
#print numbers 1 to 5

for i in range(1,6):    
    print(i)
```

```python
#for loop with list items

fruits=["mango","banana","apple"]
for i in fruits:
    print(i)
```

```python
#loop through a string

name="Python"

for letter in name:
    print(letter)

#strings are a collection of characters
```

```python
#numbers list
numbers=[10,20,30]

for n in numbers:
    print(n*2)
```

```python
#for loop with a dictionary

student={"name":"alex","age":22}

for value in student.values():
    print(value)
```

```python
student={"name":"alex","age":22}

for key in student:
    print(key)
```

```python
#looping through key-value pairs
for key,value in student.items():
    print(key,":",value)
```

```python
#print even numbers from 0-10

for i in range(0,11,2):    #start,stop,step   
    print(i)
```
