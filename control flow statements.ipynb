Conditionals(if,elif,else)

Conditionals let a program make decisions.
"If something is true-> do this. Otherwise->do that"

Basic syntax-
if condition:
    runs if condition is true
elif another_condition:
    runs if above is false and this is true
else:
    runs if all conditions are false

```python
#simple if statement
age=20
if age>=18:
    print("you can vote")
```

```python
#if-else statement

rain=False

if rain:
    print("Take umbrella")
else:
    print("don't take umbrella")
```

```python
#if-else

marks=35
if marks>=40:
    print("pass")
else:
    print("fail")
```

```python
#elif statement

weather="sunny"

if weather=="rainy":
    print("Take umbrella")
elif weather=="sunny":
    print("Take sunglasses")
else:
    print("it's a pleasant day")
```

```python
#if-elif-else

score=78

if score>=90:
    print("Grade A")
elif score>=75:
    print("Grade B")
elif score>=60:
    print("Grade C")
else:
    print("Grade D")
```

```python
#even or odd

num=8
if num%2==0:
    print("It is even")
else:
    print("it is odd")
```

```python
#login check
username="admin"
password="1234"

if username=="admin" and password=="1234":
    print("login successful")
else:
    print("login failed")
```

Structural pattern matching (match-case):

Match-case is python's switch statement

use it when:
you compare one variable against many fixed values

Baic syntax:

match value:
    case option1:
        #code
    case option2:
        #code
    case option3:
        #code
    case _:
        #default

here _ means "anything else"


```python
#menu selection
choice=int(input("Enter your choice"))

match choice:
    case 1:
        print("Add item")
    case 2:
        print("Delete item")
    case 3:
        print("Update item")
    case _:
        print("invalid choice")
```

```python
#using strings

day="Sunday"

match day:
    case "Saturday"|"Sunday":
        print("Weekend")
    case _:
        print("Weekday")
```

```python
#simple calculator option

operation="+"

match operation:
    case "+":
        print("Addition selected")                       #match ->check this value
    case "-":                                            #case  ->if it matches this
        print("Subtraction selected")                    # _    -> anything else
    case "*":
        print("Multiplication selected")
    case "/":
        print("Division selected")
    case _:
        print("invalid operation")
```

match is not for conditions like greater than or less than 
match cannot be used for a range of values either

case age>18 #invalid
case num>30 #invalid

```python
#Truthy or Falsy

#what should happen if a list or variable is empty 

items=[]

if items:
    print("items exist")
else:
    print("no items")

#python treats empty things as false
```
