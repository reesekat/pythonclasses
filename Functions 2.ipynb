Flexible arguments-
*args and **kwargs

Normally a function expects a fixed number of inputs.

def add(a,b):
    return a+b
add(4,5)

This works only if you pass exactly two values

What if you have 2 values,sometimes 3 or 10

That's why we have flexible arguments.

*args- variable poisitional arguments
(accepts any number of values,in order)

*kwargs- variable keyword arguments
(accept any number of named arguments)

*args rules-

args is just a name(can be anything)
* is mandatory
values are stored as a tuple
order matters

```python
def add(*args):
    return args+args+args
```

```python
def demo(*args):
    print(args)

demo(1,2,3,4,5,6)
```

```python
#sum any number of values

def add_all(*ar):
    total=0
    for n in ar:
        total+=n
    print(type(ar))
    return total

print(add_all(1,2))
print(add_all(1,2,3,4))
```

```python
#find maximum value
def find_max(*numbers):
    max_val=numbers[0]
    for n in numbers:
        if n>max_val:
            max_val=n
    return max_val
    
print(find_max(10,5,30,20))
print(find_max(60,70,20,80))
```

**kwargs rules-

must be key=value pairs
stored as a dictionary
order does not matter

```python
def demo(**kwargs):
    print(kwargs)

demo(name="Alex",age=22)
demo(name="Alex",age=22,salary=100000) 
#internally kwargs turns into a dictionary
```

```python
#display user-profile
def show_profile(**details):
    for key,value in details.items():
        print(key,":",value)

show_profile(name="Sam",age=23,city="Bangalore")
```

```python
#order summary

def order_summary(**order):
    print("Order summary")
    for item,detail in order.items():
        print(item,":",detail)

order_summary(product="Laptop",price=50000)
order_summary(product="phone",price=30000,warranty="2 years")
```

```python
#student registration
def register_student(**info):
    print("Student registration")
    print("--------------------")
    for key,value in info.items():
        print(key.capitalize(),":",value)


register_student(name="Asha",age=20)
register_student(name="Ravi",age=21,course="Data analytics",course_fee=22000)
```

```python
#application settings

def app_settings(**settings):
    theme=settings.get("theme","light")
    font=settings.get("font","Arial")
    notifications=settings.get("notifications",True)
    
    print("Theme:",theme)
    print("Font:",font)
    print("Notifications:",notifications)

app_settings(theme="dark")
app_settings(theme="dark",font="Helvetica",notifications=False)
```

```python
#validation function

def validate_user(**info):
    if "email" not in info:
        return "Email required"
    if "password" not in info:
        return "Password required"
    return "User validated"

validate_user(email="a@b.com",password="1234")
```

```python
#mixing normal parameters + kwargs

def send_message(to,**options):
    print("Sending message to:",to)

    if "urgent" in options:
        print("Priority:HIGH")

    if "cc" in options:
        print("CC:",options["cc"])

send_message("Alex",urgent=True)
send_message("Sam",cc="manager",urgent=True)
```
