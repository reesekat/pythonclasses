Function-
A function is a reusable code that performs a specific task.
Instead of writing the same code again and again we define it once and reuse it.

why functions matter for data analysts?

->reuse logic(cleaning,calculation)
->organize code
->avoid duplication
->make analysis readable

Defining functions-

Functions are defined using the def keyword

Syntax:
def function_name():
    code

```python
#simple function example

def greet():
    print("Hello")    #Defining a function

greet()   #calling the function
```

```python
#function with parameters

def greet(name):
    print("Hello",name)

greet("Sri")
```

Parameters and arguments

Parameters-
parameters are placeholders written inside the function

example-

def add(a,b):
    return a+b

here a and b are parameters
they don't have real values yet

think of it like these are empty boxes waiting for values

Arguments-
Arguments are the actual values passed to the function.

example-
add(2,3)  #calling  a function

here 2 and 3 are arguments
they fill the parameters a and b.

Parameters are used when defining a function
Arguments are used when calling a function

```python
#Function arguments
#positional vs keyword arguments

#positional arguments
#here order matters

def add(a,b):
    print(a+b)  
      
add(2,3)
```

```python
#keyword arguments

def add(a,b):
    print(a+b)  

add(b=3,a=2)
#here order does not matter

#when to use keyword arguments-
#when many parameters
#when meaning matters
#when writing clean,readable code
```

Return values-

what does return do?
sends a value back to the caller
ends the function
allows further computation

```python
#example 1: function with print only

def add(a,b):
    print(a+b)

result=add(2,3)
print (result)
```

```python
#function with return

def add(a,b):
    return a+b
result=add(2,3)
print(result)
```

```python
def square(n):
    return n*n

result=square(4)
print(result)

#function calculates  4*4
#return 16 sends value back
#result=16
```

```python
#code after return never runs

def test():
    print("Before return")
    return 10
    print("After return")  #never runs

x=test()
print(x)
```

```python
#returning multiple values

def calculate(a,b):
    total=a+b
    diff=a-b
    return total,diff

result=calculate(10,5)
print(result)
```

Documentation

Writing docstrings

A docstring explains:
-what the function does
-parameters
-return value

it helps:

other developers
your future self
documentation tools

Basic docstring syntax-
def function():
    """
    Description
    """

```python
#google-style docstring(beginner-friendly)

def add(a,b):
    """
    Adds two numbers.

    Args:
        a(int):First number
        b(int):Second number
    Returns:
        int:Sum of a and b
    """

    return a+b
```

```python
#numpy-style docstring(industry common)

def add(a,b):
    """
    Adds two numbers.

    Parameters
    ----------
    a : int
        First number
    b : int
        Second number

    Returns
    -------
    int
        Sum of a and b
    """
    return a + b
```

```python
#to view docstring
help(add)
```

```python
#sales summary function

def sales_summary(sales):
    total=0
    highest=sales[0]
    lowest=sales[0]

    for amount in sales:
        total=total+amount

        if amount>highest:
            highest=amount

        if amount<lowest:
            lowest=amount

    average=total/len(sales)
    return total,highest,lowest,average

summary=sales_summary([100,200,150,300])
print(summary)
```

```python
#attendance percentage

def attendance_percentage(attended,total):
    if total==0:
        return 0
    
    percentage=(attended/total)*100

    if percentage>=75:
        status="Eligible"
    else:
        status="Not eligible"

    return percentage,status

p,s=attendance_percentage(18,20)
print(p,s)
```
